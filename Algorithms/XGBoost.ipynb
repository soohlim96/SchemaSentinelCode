```python
plt.figure(figsize=(6,4))
sns.kdeplot(df["PERSON_AGE"], fill=True, bw_adjust=0.8)
plt.title("Age Distribution (Smoothed)")
plt.xlabel("Age")
plt.ylabel("Density")
plt.xlim(0, 90)
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="PERSON_SEX")
plt.title("Gender Distribution in MVCs")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.ylim(0, 300000)   # LOWER Y-AXIS MORE
plt.show()

```
